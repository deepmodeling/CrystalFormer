```python
import numpy as np
import pandas as pd
from ast import literal_eval

from pymatgen.core import Structure
from matbench_genmetrics.core.utils.plotting import plot_structures_2d
```

```python
data = pd.read_csv('./output_225_struct.csv')
cif_strings = data['cif']   # string of pymatgen structure dict

structures =[Structure.from_dict(literal_eval(cif)) for cif in cif_strings]
```

```python
np.random.seed(42)
# random pick 8 structures from the structures list
idx = np.random.choice(len(structures), 8, replace=False)
# print(idx)

selected_structures = [structures[i] for i in idx]

fig, axes = plot_structures_2d(selected_structures, 2, 4)
fig.set_size_inches(20, 8)
```
